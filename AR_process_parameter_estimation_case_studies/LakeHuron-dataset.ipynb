How one can remove the trend (i.e. de-trend) the time series 'LakeHuron' in R? You can use the code block below to check your answer.

```r
#looks like there is a down-trend
#lets remove the trend using 'detrend' function
#install.packages('pracma')
library('pracma')
data <- LakeHuron
class(as.numeric(data))
detrended <- detrend(as.numeric(data))
par(mfrow = c(4,1))
plot(data,main='LakeHuron time-series data')
plot(ts(detrended),main='detrended LakeHuron time-series data')
plot(data-mean(data),main='demeaned LakeHuron time-series data')
plot(diff(data),main='differenced LakeHuron time-series data')
```

Find the PACF of the differenced time series



```r
differenced_data <- diff(data)
par(mfrow=c(2,1))
acf(differenced_data)
acf(differenced_data,type='partial',lag.max = 30)
```

```r
acf_coeff <- acf(differenced_data,plot=F)
acf_coeff
```
