### Objective
    1. To describe what is Auto-Regressive process
    2. Simulate AR process in R
    3. Discuss qualitative features of ACF of AR process
    4. Express Random-walk as AR(1) process

### Functional form of AR(p) process

    Xt = Zt+history
    nothing but ** Current Random State + Linear combination of previous history upto lag 'p' **
    ie, Xt = Zt+phi1*Xt-1+phi2*Xt-2+....+phip*Xt-p
    where p - lags

```r
#Simulate AR(1) in R
set.seed(2016)
phi <- .4
Z <- rnorm(1000,0,1)
X <- NULL
X[1] <- Z[1]
for(i in 2:1000){
    X[i] <- Z[i]+.4*X[i-1]
}
```

```r
#plotting
par(mfrow = c(2,1))
plot(ts(X),main='Simulation of AR(1) process')
(acf(X,type='correlation',main='auto-correlation-function of AR(1)'))
```

```r
#Simulate AR(2) process of the form : X(t) = Z(t)+.7*X(t-1)+.2*X(t-2)
set.seed(2017)
X.ts <- arima.sim(list(ar=c(.7,.2)),n=1000)
par(mfrow=c(2,1))
plot(X.ts,main="AR(2) of the form X(t) = Z(t)+.7*X(t-1)+.2*X(t-2) ")
(acf(X.ts,main="ACF of AR(2)"))
```
