# *USAccDeaths* dataset

```r
#help(USAccDeaths) #Accidental Deaths in the US 1973–1978
class(USAccDeaths)
USAccDeaths
par(mfcol = c(2,1))
plot(USAccDeaths,main='Accidental Deaths in the US 1973–1978',lwd=2)
plot(USAccDeaths[13:36],type='l',main='2 years',lwd=2)
```

```r
#Time-series is definitely not stationary since there is a clear 'seasonal' trend
par(mfrow = c(2,1))
acf(USAccDeaths,main='ACF',lag.max = 24)
pacf(USAccDeaths,main='PACF',lag.max = 24)
```

```r
#let's look at the plot of the seasonal differenced data
deaths_diff <- diff(USAccDeaths,12)
par(mfrow = c(3,1))
plot(deaths_diff,main='Accidental Deaths in the US 1973–1978',lwd=2)
acf(deaths_diff,main='ACF',lag.max = 24)
pacf(deaths_diff,main='PACF',lag.max = 24)
Box.test(deaths_diff,type='Ljung-Box',lag=log(length(deaths_diff)))
```

```r
#install.packages('LSTS')
library('LSTS')
#m <- Box.Ljung.Test(diff(USAccDeaths,12), lag = 30, main = 'plot')
m <- ts.diag(diff(USAccDeaths,12))
```

```r
acData <- diff(diff(USAccDeaths),12)
par(mfrow = c(3,1))
plot(acData,main='seasonal and non-seasonal differenced data',lwd=2)
acf(acData,main='ACF',lag.max = 24)
pacf(acData,main='PACF',lag.max = 24)
Box.test(acData,type='Ljung-Box',lag=log(length(acData)))
```
