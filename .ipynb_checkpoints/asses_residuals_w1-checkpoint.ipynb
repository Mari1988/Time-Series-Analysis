## Asses the normality of residuals
### we have seen that simple straight line was not enough to capture co2 concentration over time because of oscillatory nature in the data

```r
plot(co2,main='Atmospheric concentrations of CO2')
co2_model <- lm(co2 ~ time(co2)) #co2 concentration as a function of time
abline(co2_model) #abline is the intercept-slope line
#Why should the error be normal
#As we know the average value of residuals is zero, most of the residuals are expected to centered around zero
```

#### plotting histogram of residuals to asses the normality is not a good approach due to less number of observations

```r
hist(resid(co2_model),xlab='residuals',main='histogram of residuals')
```

### qqnorm () function plots the observed (empirical) line to the reference line
### qqline() gives the reference (theoretical) line for the normality test

```r
qqnorm(resid(co2_model))
qqline(resid(co2_model))
#We can see that extreme values are off from the reference line
```

## time plot of the residuals (residuals as a function of time)

```r
plot(resid(co2_model) ~ time(co2))
```

### above residual plot shows that,
1. for the first part there is an underprediction
2. for the middle part there is an overprediction
3. the final part again underprediction
#### however the oscillatory part in residual is not very obvious from this view. so let's zoom in

```r
plot(resid(co2_model) ~ time(co2),xlim = c(1960,1965),main='zoomed-residuals')
```

### it is very clear from the above plot that oscillatory parts are not picked up by the linear model as a function of time
