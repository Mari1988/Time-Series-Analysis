# fit a linear model to a set of ordered pairs

```r
#Atmospheric concentrations of CO2
#help(co2)
plot(co2,main='Atmospheric concentrations of CO2')
#certainly linear model doesnt look like an option here. Because straight line will not capture the essence of 
#trend/seasonal(oscillations) variations
```

## Concerns in fitting just a linear model

Fitting a linear model is not a problem here however we need to see the error componenet if it makes sense.
Here in the above example, response variable is co2 concentration level
which can be expressed as <br>
Y_i = beta0+beta_i*x_i + error_i <br>
The error term could be due to <br>
1. measurement error <br>
2. lack of knowledge of other important factors , etc <br>


## Assumptions about error term from linear model

1. error terms should be normally distributed, zero mean on average
2. error term have constant variance (ie, they are homoscedastic)
3. each observation is uncorrelated (no autocorrelation) with others (ie, they are white) 

```r
#linear model
co2_model <- lm(co2 ~ time(co2)) #co2 concentration as a function of time
summary(co2_model)
```

```r
head(time(co2))
```

```r
plot(co2,main='Atmospheric concentrations of CO2')
abline(co2_model) #abline is the intercept-slope line
```

#### it is very clear from the plot that 'straight line' is not the best fit even to capture the trend
