### Objective

    1. Obtain PACF (Partial Auto-Correlation Function) using ACF function
    2. Estimate likely order of an AR(p) process using PACF function
    3. Estimate the coefficients of AR(p) process using ar() function
    

##### determining order of a MA(q) process is quite straightforward -> we just need to look at the ACF plot and determine

#### How do we determine the order of an AR ???????

#### Recall of functional form of AR(p)

$X_t = Z_t+\phi_1X_{t-1}+\phi_2X_{t-2}+...........+\phi_pX_{t-p}$

```r
#let's try to simulate the same in R
set.seed(20180820)
phi1=.6
phi2=.2
data.ts <- arima.sim(list(ar=c(phi1,phi2)),n=500)
```

```r
#plot the results
par(mfrow = c(3,1))
plot(data.ts,main='AR(2)')
acf(data.ts,main='ACF of AR(2)')
acf(data.ts,type='partial',main='PACF of AR(2)')
```

#### note from above plots
    1. It's hard to know the lag of AR process from ACF plot
    2. However , PACF plot suggests (peak at lag-1 and lag-2) the lag for us. 
    
**This is incredible !!!!!!!!!!! **

```r
#let's try to simulate the AR(3)
set.seed(20180820)
phi1=.9
phi2=-.6
phi3=-.3
data.ts <- arima.sim(list(ar=c(phi1,phi2,phi3)),n=500)

#plot the results
par(mfrow = c(3,1))
plot(data.ts,main='AR(3)')
acf(data.ts,main='ACF of AR(3)')
acf(data.ts,type='partial',main='PACF of AR(3)')
```

#### We could see 3 statistically significant spikes

### Let's analyze the real dataset called Beveridge wheat price index, 1500-1869
https://datamarket.com/data/set/22ts/beveridge-wheat-price-index-1500-1869#!ds=22ts&display=line


```r
beveridge <- read.csv("beveridge-wheat-price-index-1500.csv",stringsAsFactors=F)
colnames(beveridge) <- c("Year","WheatPrice")
ts.wp <- ts(beveridge$WheatPrice,start=1500)
plot(ts.wp,xlab="Year",ylab="Price",main="wheat-price from 1500 to 1869 (yearly data)")
wp.ma <- filter(ts.wp,rep(1/31,31),sides=2) #moving average of 31 years
lines(wp.ma,col="red") #superimpose on the previous plot
#MA process does nice job in smoothening the fluctuations and at the same time it captured the essence of the trend
```

```r
#scale the original beveridge time-series by the smoothened beveridge ts -- This is done to introduce the stationary process
Y <- ts.wp/wp.ma
par(mfrow=c(3,1))
plot(Y,ylab = "scaled-price",main="Transformed Beveridge wheat price")
acf(na.omit(Y),main="ACF of Transformed Beveridge data")
acf(na.omit(Y),type="partial",main="PACF of Transformed Beveridge data")
```

#### Interesting observations from above plots

    1. ACF plot suggests lag of '1' -- one significant peak
    2. PACF plot suggests lag of '2' -- two significant peak
    
Which one to trust ? Let's run AR process and figure out the answer

```r
ar(na.omit(Y),order.max = 5)
```

### AR function is also suggesting AR(2) process. So the clear winner is 'PACF'
